```python
df
```

```python
X2=df[['_unit_state', '_trusted_judgments',
       '_last_judgment_at', 'gender:confidence',
       'profile_yn:confidence', 'created', 'fav_number', 'link_color', 'name',
       'profileimage', 'retweet_count', 'sidebar_color', 'tweet_count',
       'tweet_created','tweet_location', 'user_timezone']].values
Y2=df[["gender"]].values
```

```python
from sklearn.model_selection import train_test_split
X2_train,X2_test,Y2_train,Y2_test=train_test_split(X2,Y2,train_size=0.8)
X2_train.shape,X2_test.shape,Y2_train.shape,Y2_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=300)
rfc.fit(X2_train,Y2_train)
```

```python
Y2_pred=rfc.predict(X2_test)
Y2_pred
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(Y2_test,Y2_pred)
a
```

# accuracy with varying no of estimators :

```python
rf={}
for i in range(100,1000,100):
    rfc=RandomForestClassifier(n_estimators=i)
    rfc.fit(X2_train,Y2_train)
    Y2_pred=rfc.predict(X2_test)
    a=accuracy_score(Y2_test,Y2_pred)
    rf[i]=a
print(rf)    
```

```python
X3=df[['_unit_state', '_trusted_judgments',
       '_last_judgment_at', 'gender:confidence',
       'profile_yn:confidence','fav_number','name',"link_color",
       'profileimage', 'retweet_count',"sidebar_color",'tweet_count','tweet_created',
       'tweet_id', 'tweet_location', 'user_timezone']].values
Y3=df[["gender"]].values
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
X3_train,X3_test,Y3_train,Y3_test=train_test_split(X3,Y3,train_size=0.8)
X3_train.shape,X3_test.shape,Y3_train.shape,Y3_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=4,metric="euclidean")
knn
```

```python
knn.fit(X3_train,Y3_train)
```

```python
Y3_pred=knn.predict(X3_test)
Y3_pred
```

```python
from sklearn.metrics import accuracy_score
a=accuracy_score(Y3_pred,Y3_test)
a
```

# Generating accuracy with different no of neighbors and stroing accuracy in dictionary.

```python
accu={}
for i in range(2,15):
    knn= KNeighborsClassifier(n_neighbors=i,metric="euclidean")
    knn.fit(X3_train,Y3_train)
    Y3_pred=knn.predict(X3_test)
    a=accuracy_score(Y3_pred,Y3_test)
    accu[i]=a
print(accu)    
    
    
```
